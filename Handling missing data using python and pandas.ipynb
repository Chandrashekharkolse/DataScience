### Calculate mean values

```python
age_mean = df['Age in Yrs'].mean()
weight_mean = df['Weight in Kgs'].mean()
Age_Company_mean = df["Age in Company"].mean()
salary_mean = df["Salary"].mean()
print(age_mean, weight_mean, Age_Company_mean, salary_mean)
```

### Calculate meadian values

```python
age_median = df['Age in Yrs'].median()
weight_median = df['Weight in Kgs'].median()
Age_Company_median = df["Age in Company"].median()
salary_median = df["Salary"].median()
print(age_median, weight_median, Age_Company_median, salary_median)
```

#### Fill null values with mean of respective column values

```python
df['Age in Yrs'] = df['Age in Yrs'].fillna(df['Age in Yrs'].mean())
df["Weight in Kgs"] = df["Weight in Kgs"].fillna(df['Weight in Kgs'].mean())
df["Age in Company"] = df["Age in Company"].fillna(df["Age in Company"].mean())
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
```

```python
df
```

```python
df = pd.read_csv("Emp_Records.csv")
df
```

#### Fill null values with median of respective column values

```python
df['Age in Yrs'] = df['Age in Yrs'].fillna(df['Age in Yrs'].median())
df["Weight in Kgs"] = df["Weight in Kgs"].fillna(df['Weight in Kgs'].median())
df["Age in Company"] = df["Age in Company"].fillna(df["Age in Company"].median())
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```
